```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Criando o DataFrame com 17 estudantes de Moçambique
dados = {
    'nome': [
        'Aminata', 'Carlos', 'Fatima', 'João', 'Isabel', 'Hassan', 'Lúcia', 'Mário',
        'Nadia', 'Pedro', 'Rita', 'Samuel', 'Tatiana', 'Vítor', 'Zainab', 'António', 'Beatriz'
    ],
    'curso': [
        'Engenharia Informática', 'Medicina', 'Engenharia Civil', 'Direito', 'Enfermagem',
        'Arquitetura', 'Engenharia Informática', 'Medicina', 'Direito', 'Enfermagem',
        'Engenharia Civil', 'Arquitetura', 'Engenharia Informática', 'Medicina', 'Direito',
        'Enfermagem', 'Engenharia Civil'
    ],
    'idade': [
        18, 20, 19, 21, 20, 22, 19, 21, 18, 20, 19, 22, 18, 21, 20, 19, 22
    ],
    'nota': [
        82, 91, 76, 88, 85, 79, 93, 87, 81, 90, 77, 84, 92, 89, 78, 86, 80
    ]
}

df = pd.DataFrame(dados)

# Mostrando o DataFrame
print("DataFrame:\n", df)

# Calculando média, nota mínima e nota máxima
media_nota = df['nota'].mean()
nota_minima = df['nota'].min()
nota_maxima = df['nota'].max()

print(f"\nMédia das notas: {media_nota}")
print(f"Nota mínima: {nota_minima}")
print(f"Nota máxima: {nota_maxima}")

# Usando describe
print("\nDescrição estatística do DataFrame:")
print(df.describe())

# Criando gráfico de barras
plt.figure(figsize=(12,6))
plt.bar(df['nome'], df['nota'], color='orange')
plt.xlabel('Nome dos Estudantes')
plt.ylabel('Notas')
plt.title('Notas dos Estudantes em Moçambique')
plt.xticks(rotation=45)  # Rotaciona os nomes para melhor visualização
plt.show()
```
